```python
games_nba = games_nba[['GAME_DATE_EST','GAME_ID','GAME_STATUS_TEXT','TEAM_ID_home',
                      'TEAM_ID_away','PTS_home','PTS_away','HOME_TEAM_WINS']]
games_nba.head()
```

```python
#games_nba.loc[:, 'GAME_DATE_EST'] = pd.to_datetime(games_nba['GAME_DATE_EST'])
games_nba['GAME_DATE_EST'] = pd.to_datetime(games_nba['GAME_DATE_EST'])
```

```python
games_nba.info()
```

```python
games_nba.head()
```

```python
games_nba['year'] = games_nba['GAME_DATE_EST'].dt.year
games_nba.head()
```

```python
# Cargar el conjunto de teams
teams_nba.head()
```

```python
teams_nba = teams_nba[['TEAM_ID','CITY','NICKNAME']]
teams_nba.head()
```

```python
home_games = pd.merge(games_nba,teams_nba, left_on='TEAM_ID_home', right_on = 'TEAM_ID',
                     how='inner')
home_games.head()
```

```python
home_games = home_games.rename(columns={
    'CITY':'city_home',
    'NICKNAME':'nickname_home'
})
home_games.head()
```

```python
both_games = pd.merge(home_games,teams_nba, left_on='TEAM_ID_away', right_on = 'TEAM_ID',
                     how='inner')
both_games.head()
```

```python
both_games = both_games.rename(columns={
    'CITY':'city_away',
    'NICKNAME':'nickname_away'
})
both_games.head()
```

```python
both_games = both_games.drop(columns=[
    'TEAM_ID_home', 'TEAM_ID_away',
    'TEAM_ID_x','TEAM_ID_y'
])
both_games.head()
```

```python
games = both_games.copy()
games.columns = games.columns.str.lower()
games.head()
```

```python
games['pts_total'] = games['pts_home'] + games['pts_away']
games.head()
```

```python
#Encontramos los 5 partidos con mas puntos totales.
games.nlargest(n=5, columns ='pts_total')
```

```python
#Encontramos los 5 partidos con menos puntos totales.
games.nsmallest(n=5, columns ='pts_total')
```

## Cantidad de puntos por equipo cuando juegan de local o de visita.

```python
home_wins = games[games['home_team_wins'] == 1]
result = home_wins.groupby('nickname_home')['pts_home'].aggregate(np.mean).reset_index().sort_values('pts_home')
```

```python
plt.figure(figsize=(20,6))
ax = sns.barplot(data=home_wins,x='nickname_home',y='pts_home', order=result['nickname_home'],
                errorbar = None)
ax.bar_label(ax.containers[-1], fmt='%.0f',label_type='center',rotation=45)
plt.xticks(rotation=75)
plt.show()
```

```python
away_wins = games[games['home_team_wins'] == 0]
result = away_wins.groupby('nickname_away')['pts_away'].aggregate(np.mean).reset_index().sort_values('pts_away')
plt.figure(figsize=(20,6))
ax = sns.barplot(data=away_wins,x='nickname_away',y='pts_away', order=result['nickname_away'],
                errorbar = None)
ax.bar_label(ax.containers[-1], fmt='%.0f',label_type='center',rotation=45)
plt.xticks(rotation=75)
plt.show()
```
